```python
breast_cancer_mean_db_final = breast_cancer_mean_db.ix[:,[1,2,3,6,7,8,10,11]]
```

```python
breast_cancer_mean_db_final.head()
```

```python
from sklearn.cross_validation import train_test_split

def split_breast_cancer_data(feature_cols, target):
    X = breast_cancer_mean_db_final[feature_cols]
    y = breast_cancer_mean_db_final[target]
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9)
    return X_train, X_test, y_train, y_test
```

```python
feature_cols = ['radius_mean', 'texture_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'symmetry_mean', 'fractal_dimension_mean']
target = 'diagnosis'
X_train, X_test, y_train, y_test = split_breast_cancer_data(feature_cols, target)
```

```python
y_test
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics 
metrics.accuracy_score(y_test, y_pred)
```

```python
from sklearn import preprocessing
lb = preprocessing.LabelBinarizer()
lb.fit(y_pred)
lb.fit(y_test)
```
